```python
mean=[]
defect=[]
divide=[15,15,15]
for a in range(len(data['care_areas'])):
  for i in range(999-data["care_areas"][a]['top_left']['y'],999-data["care_areas"][a]['bottom_right']['y']):
    t=[]
    for j in range(data["care_areas"][a]['top_left']['x'],data["care_areas"][a]['bottom_right']['x']):
      mean=np.array([0,0,0])
      for image in wafer:
        mean=np.add(image[i][j],mean)
      mean=np.divide(mean,divide)
      t.append(mean)
      c=[]
      for image in range(len(wafer)):
        c.append(np.absolute(np.subtract(wafer[image][i][j],mean))[0])
      n=[i for i in c if c.count(i) <15//2]
      for image in range(len(wafer)):
        for k in n:
          if (np.absolute(np.subtract(wafer[image][i][j],mean))==np.array([k,k,k])).all():
            defect.append([image+1,j,999-i])
exclusion=[]
for a in range(len(data['exclusion_zones'])):
  for i in range(999-data['exclusion_zones'][a]['top_left']['y'],999-data['exclusion_zones'][a]['bottom_right']['y']):
    t=[]
    for j in range(data['exclusion_zones'][a]['top_left']['x'],data['exclusion_zones'][a]['bottom_right']['x']):
      mean=np.array([0,0,0])
      for image in wafer:
        mean=np.add(image[i][j],mean)
      mean=np.divide(mean,divide)
      t.append(mean)
      c=[]
      for image in range(len(wafer)):
        c.append(np.absolute(np.subtract(wafer[image][i][j],mean))[0])
      n=[i for i in c if c.count(i) < 15//2]
      for image in range(len(wafer)):
        for k in n:
          if (np.absolute(np.subtract(wafer[image][i][j],mean))==np.array([k,k,k])).all():
            exclusion.append([image+1,j,999-i])

print(exclusion)

```
